```python
#IMPORT & LOAD DATA
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load clean data
df = pd.read_csv("D:/customer-segmentation-unsupervised/data/processed/clean_data.csv")

print("Data Loaded ")
df.head()
```

```python
#BASIC OVERVIEW
print("Shape:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nEvent Types:")
print(df["event_type"].value_counts())
```

```python
#VISUALIZATION SECTION
#EVENT TYPE DISTRIBUTION
plt.figure()
sns.countplot(x="event_type", data=df)
plt.title("Event Type Distribution")
plt.xlabel("Event Type")
plt.ylabel("Count")
plt.show()
```

```python
# PRICE DISTRIBUTION
plt.figure()
sns.histplot(df["price"], bins=50)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# TOP PRODUCT CATEGORIES
top_cat = df["category_code"].value_counts().head(10)

plt.figure()
top_cat.plot(kind="bar")
plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# USER ACTIVITY BY HOUR
plt.figure()
sns.countplot(x="hour", data=df)
plt.title("User Activity by Hour")
plt.xlabel("Hour")
plt.ylabel("Count")
plt.show()
```

```python
# DAY-WISE ACTIVITY (BONUS)
plt.figure()
sns.countplot(x="day_of_week", data=df,
              order=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"])
plt.title("Activity by Day of Week")
plt.xticks(rotation=45)
plt.show()
```

```python
# PRICE VS EVENT TYPE
plt.figure()
sns.boxplot(x="event_type", y="price", data=df)
plt.title("Price Distribution by Event Type")
plt.show()
```

```python
# TOP USERS (FREQUENCY)
top_users = df["user_id"].value_counts().head(10)

plt.figure()
top_users.plot(kind="bar")
plt.title("Top Active Users")
plt.xlabel("User ID")
plt.ylabel("Activity Count")
plt.show()
```
